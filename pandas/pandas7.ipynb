```python
df1 = pd.read_csv('Data09.csv', encoding='cp949')
print(df1.shape)
df1.head(10)
```

```python
df1.columns
```

```python
df1.drop(columns=['순번', '상품코드'])
```

```python
df1.drop(columns=['순번', '상품코드']).stack()
```

```python
stack_data = df1.drop(columns=['순번', '상품코드']).set_index('상품명').stack() # 상품명을 인덱스로 보내야만 상품명이 보이지 이렇게 인덱스를 설정안하면 상품명이 뭔지 모르게 된다.
```

```python
pd.DataFrame(stack_data).reset_index()
```
