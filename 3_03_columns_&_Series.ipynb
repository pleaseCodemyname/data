## 2. 데이터셋 설명
- 데이터셋은 가상의 온라인 상점의 고객 주문 내역을 포함하고 있다. 데이터셋의 컬럼은
  - CustomerID : 고객의 고유 ID
  - OrderDate : 주문날짜
  - Product : 구매한 상품명
  - Quantity: 구매한 상품의 수량
  - Price : 상품의 가격
- 예제
  - 컬럼 이름 변경: CustomerID 컬럼의 이름 ID로 변경
  - 컬럼 삭제: OrderData 컬럼 삭제
  - 새 컬럼 추가: 각 주문의 총 금액을 나타내는 TotalAmount 새 컬럼 추가. 이 컬럼은 Quantity와 Price의 곱으로 계산

````python
import pandas as pd

# 예제
data = {
    'CustomerID': [101, 102, 103, 104],
    'OrderDate': ['2021-01-01', '2021-01-02', '2021-01-03', '2021-01-04'],
    'Product': ['Apple', 'Banana', 'Cherry', 'Date'],
    'Quantity' : [5, 3, 4, 2],
    'Price' : [100, 200, 250, 50]
}
df = pd.DataFrame(data)
````

````python
df.rename(columns={'CustomerID': 'ID'}, inplace = True) # inplace = True를 해야 바뀐 값이 반영
df
df.drop(labels = 'OrderDate', axis=1, inplace=True) # drop(labels='', axis=1 (column), inplace=True)
df
````

````python
df['TotalAmount'] = df.Quantity * df.Price
````
