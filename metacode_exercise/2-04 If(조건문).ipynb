# Boolean
- 참(True)과 거짓(False)만을 나타내는 데이터 타입
  - 비교연산자
    - ==(Equal to): 두 값이 ***같은지*** 비교. 같으면 참(True) 반환
    - !=(Not equal to): 두 값이  ***다른지*** 비교. 다르면 참(True) 반환
    - >(Greater than): 왼쪽 값이 오른쪽 값보다 ***큰지*** 비교. 크면 참을 반환
    - <(Less than): 왼쪽 값이 오른쪽 값보다 ***작은지*** 비교. 작으면 참을 반환
    - >=(Greater than or equal to): 왼쪽 값이 오른쪽 값보다 ***크거나 같은지*** 비교. 크거나 같으면 참을 반환
    - <=(Less than or eqaul to): 왼쪽 값이 오른쪽 값보다 ***작거나 같은지*** 비교. 작거나 같으면 참을 반환.


# IF(조건문)
- 조건문은 프로그래밍에서 "만약 이 조건이 참이라면 이 코드를 실행"라는 의미를 가진 구조


```
if 조건:
  # 조건이 참(True)일 때 실행되는 코드
else:
  # 조건이 거짓(False)일 때 실행되는 코드

```

````python
if 1!=1: # 1 > 2 | 1 < 2 | True | False | true |false
  print('맞다')
else:
  print('틀리다')
````

# elif

```
if 조건1:
  # 조건1이 참(True)일 때 실행
elif 조건2:
  # 조건1이 거짓(False)이고, 조건2가 참(True)일 때 실행
else:
  # 모든 조건이 거짓(False)일 때 실행
```




````python
if 1==2 : # 조건1
  print("조건 1이 맞다")
elif 1!=1: # 조건2
  print("조건 1은 틀리고 조건 2가 맞다")
else:
  print("둘다 틀리다")
````

1. 문제: Python 함수 compare를 작성하세요. 이 함수는 두 개의 매개변수 a와 b를 받아, 다음 조건에 따라 결과를 반환.
  - 만약 a가 b보다 크다면, "a가 더 크다."를 반환합니다.
  - 만약 b가 a보다 크다면, "b가 더 크다."를 반환합니다.
  - 만약 a와 b가 같다면, "a와 b는 같다."를 반환합니다.

````python
def compare(a ,b):
  if a > b:
    output = f'a가 더 크다.'
  elif b > a:
    output = f'b가 더 크다.'
  else:
    output = f'a와 b는 같다.'
  return output
````

````python
compare(1, 2)
````
