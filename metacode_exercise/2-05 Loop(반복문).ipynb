# For Loop
- for 루프는 주로 시퀀스(리스트, 문자열 등 순회할 대상)를 순회하며, 각 원소에 대해 루프 내부의 코드를 실행


```
for 변수 in 시퀀스:
  # 실행할 코드
```



````python
# 리스트란: 순서가 있는 아이템들의 모음 [](대괄호)로 표현
my_list = [1, 2, 3, 4]
````

````python
# 0에서 부터 시작(왼쪽) , -1 부터 시작 (오른쪽)
my_list[-1]
````

````python
seq = 0
for i in [1, 10, 100, 1000]:
  seq = seq + 1 # seq += 1 과 동일
  print('---start')
  print(f'시퀀스: {seq}, 아이템: {i}')
  print( '---end---')
  print(' ')
# https://pythontutor.com/render.html#mode=edit
````

# While Loop
- 조건이 참(True)인 동안 반복적으로 코드 블록을 실행하는 데 사용


```
while 조건:
  # 실행할 코드
```

````python
i = 0
j = 1
while i < 5: # i=1 True
  bool = i < 5 # True | True
  i = i + 1 # 1
  j = j*10 # 10
  print(f'i:{i}, Boolean: {bool}, j:{j}') # i: 1, Boolean: True, j:10
print(f'boolean: {i < 5}') # boolean: true
````

# Exercise
- 1부터 10까지의 숫자 중에서, for 루프를 사용하여 모든 홀수의 합을 계산. 동일한 범위에서 while 루프를 사용하여 모든 짝수의 합을 계산.

````python
# for 루프를 사용하여 홀수의 합 계산
sum_odd_for = 0
for i in range(1, 11): # 1부터 10까지의 범위
  if i % 2 != 0: # 홀수인지 확인
    sum_odd_for += i # sum_odd_for = sum_odd_for + i 와 같음
print(sum_odd_for)
````

````python
sum_even_while = 0 # 초기값 0으로 설정
i = 1
while i <= 10:
  if i % 2 == 0:
    sum_even_while += i
  i += 1# i 값을 증가
print(sum_even_while)
````

````python
print(f'홀수, {sum_odd_for}')
print(f'짝수, {sum_even_while}')
````
