# 함수의 구조


```
def 함수이름(매개변수, 매개변수2, ..., 매개변수 N):

    """
    문서화 문자열 (선택사항) : 함수의 목적과 매개변수, 반환값 등에 대한 설명
    """

    # 함수의 로직, 계산, 처리 등을 수행하는 코드

    return 반환값 # 함수가 결과를 반환. 반환값이 없는 경우, return은 생략 가능

```
- 함수: 특정 작업을 수행하는 코드의 집합으로, 재사용 가능
  - Parameter(매개변수): 함수의 입력값을 지정하기 위한 변수

2x + 3 = f(x) <br/>
x  f(x) <br/>
1   5  <br/>
2   7  <br/>
3   9  <br/>
4   11  <br/>
5   13 <br/>





````python
# ^함수 설정
def operation1(x):
  '''
  2x + 3 연산을 수행
  x = int
  '''
  result = (2 * x) + 3
  return result
````

````python
# 함수 설명
help(operation1)
````

````python
print(operation1(1))
````

1. 함수 만들기
  - 함수이름: greet
  - 설명: 함수 이름을 매개변수로 받아 "Hello, [name]!"형식으로 인사말을 반환하는 함수 작성
  - 예시: ***greet("Alice")***는 "Hello, Alice!"를 반환해야 합니다.

````python
def greet(name):
  ex = f'Hello, {name}!'
  return ex
greet("Alice")
````
