# Tabular Data & Data type
- Tabular Data: Column, Row, Index, Value로 구성된 직사각형 형태의 데이터 테이불
- 행(Row): 데이터 테이블에서 가로 방향 데이터 집합. 하나의 행은 ***레코드***
- 열(Column): 데이터 테이블에서 세로 방향 데이터 집합. 하나의 열은 ***속성***
- 값(Value): ***행과 열이 교차***하는 지점에 위치한 개별 데이터.
- 인덱스(Index): ***행을 구별***하는 데 사용되는 고유한 식별자



````python
# .shape: 행, 열 수 표시.
# .columns: 열 이름 나열.
# .values: 데이터를 배열로 반환.
# .index: 행 인덱스 반환.
````

# Data Type

1. **Categorical Data** (범주형 데이터): 제한된 수의 ***범주***로 나눌 수 있는 데이터. ex) 성별, 혈액형, 학년
2. **Numeric Data** (수치데이터): 수치로 표현되며 ***계산이 가능***한 데이터
- Binary Data (이진 데이터): ***'0'과 '1'로만*** 구성된 데이터. ex) 합격 여부, 물품 구매 여부 등이 있음, 이는 경우에 따라 범주형 데이터로도 분류.
- Ratio Data (비율 데이터): 값들 사이의 ***비율 비교***가 가능한 데이터. ex) 키, 몸무게, 가격
