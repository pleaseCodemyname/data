```python
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rc('font', family='Malgun Gothic')
```

```python
df1=pd.read_csv('Data03.csv')
df1.head()
```

# 다변수

* X : 연속 / Y: 연속
    - 산점도 / 회귀선 / 상관분석 

```python
sns.scatterplot(data=df1, x='age', y="Mileage") #scatterplot: 산점도
```

```python
sns.scatterplot(data=df1, x='age', y='Mileage', hue='type_of_contract')
```

```python
sns.scatterplot(data=df1, x='age', y='Mileage', hue='type_of_contract',size='Term')
```

```python
sns.lmplot(data=df1, x='age', y='Mileage', hue='product') # lm: linear model: 선형회귀식
```

```python
sns.lmplot(data=df1, x='age', y='amount', hue='product')
```

```python
sns.jointplot(data=df1, x='age', y='amount') # 가운데는 산점도 , 바깥에는 Mileage 분포도, age분포도 표현가능 by jointplot 사용
```

```python
sns.jointplot(data=df1, x='age', y='amount', hue='type_of_contract2')
```

```python
sns.pairplot(df1) # df1 모든 연속형 변수에 대해서 산점도와 분포를 그려줌
```

```python
sns.pairplot(df1[['age','Mileage', 'amount', 'product']],hue='product')
```

```python
df1.corr() # correlation (0~1사이의 값) 1로 갈수록 양의 상관관계, -1로 갈수록 음의 상관관계, 0에 갈수록 데이터가 난잡하게 펼쳐져있다
```
